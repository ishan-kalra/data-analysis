```python
df_apps.dtypes
indexes_to_be_dropped = list(df_apps[df_apps['Size'] == 'Varies with device'].index)
df_apps.drop(indexes_to_be_dropped, inplace=True)
#df_apps.drop[index=indexes_to_be_dropped, axis=0]
```

```python
df_apps.dtypes
```

```python
#writing a function to settle app sizes with KB
def settle_sizes(df):
    pattern = "k$"
    x = re.search(pattern, df)
    if x:
        df = df.rstrip('k')
    else:
        pass
    return df
df_apps['Size'] = df_apps['Size'].map(settle_sizes)
```

```python
df_apps.drop([10472], inplace=True)
```

```python
def format_string(df):
    pattern = "[,]"
    x = re.search(pattern, df)
    if x:
        df = df.replace(',', '')
    else:
        pass
    return df
df_apps['Installs'] = df_apps['Installs'].map(format_string)
```

```python
df_apps['Size'] = df_apps['Size'].str.strip('M')
df_apps['Installs'] = df_apps['Installs'].str.strip('+')
df_apps['Price'] = df_apps['Price'].str.lstrip('$')
df_apps.drop(['Last Updated', 'Current Ver', 'Android Ver'], axis=1, inplace=True )
df_apps_cleaned = df_apps.astype(dtype={'Rating' : np.float64, 'Size' : np.float64, 'Installs' : np.uint64}, inplace=True)
df_apps_cleaned.head()
```
